```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
single_df = pd.read_excel("sorted_heat_genes.coverage.xlsx",sheet_name="Sheet9",header=0,index_col=0)
single_df
```

```python
all_df = pd.read_excel("sorted_heat_genes.coverage.xlsx",sheet_name="filter_by_2x",header=0)
all_df
```

```python
all_df.shape
```

```python
## calculate mean for all sample for single copy gene
col_means = single_df.mean(axis=0)
col_means
```

```python
# Divide each col/gene by its mean value
normalized_single_df = single_df.div(col_means,axis=1)
normalized_single_df
```

```python
row_means = normalized_single_df.mean(axis=1)
row_means
```
